```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 假设你的数据集路径
data_dir = r"E:\github复现\resnet实现项目test\data\train"

# 添加以下代码计算均值和标准差
# 自动计算均值和标准差
train_dataset = torchvision.datasets.ImageFolder(
    root=train_dir, 
    transform=transforms.Compose([transforms.ToTensor()])
 )

loader = torch.utils.data.DataLoader(
    train_dataset, 
    batch_size=32, 
    shuffle=False

mean = 0.
std = 0.
for images, _ in loader:
    batch_samples = images.size(0)
    images = images.view(batch_samples, images.size(1), -1)
    mean += images.mean(2).sum(0)
    std += images.std(2).sum(0)
mean /= len(loader.train_dataset)
std /= len(loader.train_dataset)
)   

#这里不能使用数据加载器 因为训练集中的图片并不一样大 使用loader进行批次处理需要一样大的图片 
#我们在后续预处理进行裁剪之后可以这样去做
mean = 0.
std = 0.
for i in range(len(train_dataset)):
    image, _ = train_dataset[i]
    image = image.view(image.size(0), -1)
    mean += image.mean(1)
    std += image.std(1)

mean /= len(train_dataset)
std /= len(train_dataset)

#输出train_dataset的长度，即图片数量 来查看我们是否定义正确
print(len(train_dataset)) 
mean = 0.
std = 0.
for i in range(len(train_dataset)):
    image, _ = train_dataset[i]
    image = image.view(image.size(0), -1) #-1表示自动计算展平后的维度。
    mean += image.mean(1)
    std += image.std(1)

mean /= len(train_dataset)
std /= len(train_dataset)
print(mean, std)
```
